## Exercise 16: Functional Approach - Histogram

In statistics it is essential to be aware of the distribution of continuous variables prior to running any type of analysis. To display the distribution, we will use a histogram. Histogra, display the distribution by the bin for a given array

```python
'''To demonstrate the creation of a histogram, we will generate an array of 100 normally distributed values
with a mean of 0 and standard deviation of 0.1'''

y = np.random.normal(loc = 0, scale = 0.1, size = 100)
plt.hist(y, bins = 20)
plt.xlabel('y value')
plt.ylabel('frequency')
plt.show()
```

We look at the histogram, we often determine whether the distribution is normal. Sometimes a distribution may appear normal but it is not, and sometimes a distribution may appear not normal but it is normal. 

There is a test for normality called the **Shapiro-Wilk test** .
The null hypothesis for the Shapiro-Wilk test is that the data is normally distributed. 

Thus, a p-value <0.05 indicates a non-normal distribution while a p-value> 0.05 indicates a normal distribution. 

We will use the results from the shapiro-wilk test to create a programmatic title communicating the reader whether the distribution is normal or not.

```python
'''using the tuple unpacking to store the w-statistic and p-value from the shapiro-wilk test into shap-w and shap-p objects
 respectively'''

from scipy.stats import shapiro
y = np.random.normal(loc = 0, scale = 0.1, size = 100)
plt.hist(y, bins = 20)
plt.xlabel('y value')
plt.ylabel('frequency')
shap_w, shap_p = shapiro(y)
print("p-value: ",shap_p)

if shap_p > 0.05:
    normal_YN = 'Fail to reject the null hypothesis. Data is normally distributed'
else:
    normal_YN = 'Null hypothesis is rejected. Data is not distributed normally'
plt.title(normal_YN)
plt.show()
```

## Exercise 17: Functional Approach -Box and Whisker Plot

```python
y = np.random.normal(loc = 0, scale = 0.1, size = 100)
import matplotlib.pyplot as plt
plt.boxplot(y)
plt.show()
```

The plot displays a box that represents the interquartile range(IQR). The top of the box is the 25th percentile (i.e Q1) while the bottom of the box is the 75th percentile (that is Q3).
The orange line is the median. The 2 lines extending above and beyond the box are the whiskers. The top of the upper whisker is the "maximum" value which is caclulated using Q1 - 1.5*IQR. 

The bottom of the lower whisker is the "minimum"value which is calculated by Q3 + 1.5*IQR
Outliers are displayed as dots above the "maximum" whisker and below the "minimum" whisker

```python
from scipy.stats import shapiro

shap_w, shap_p = shapiro(y)
```

```python
from scipy.stats import zscore
y_z_score = zscore(y)
```

```python
y_z_score
```

```python
'''Iterate through y_z_score array to find the number of outlier using the following code'''

num_outlier = 0

for i in range(len(y_z_score)):
    if abs(y_z_score[i]) >=3:
        num_outlier+=1
'''The number of outliers are 0 because the ayyar y was generated to be normally distributed, We can expect there
to be no outliers in the data'''
print(num_outlier)
```

## Consolidating the above code

```python
'''Generate a title, that communicates that whether the data as well as the outliers are normally distribute or not
If the shap_p is greater than 0.05, the data is normally distributed. If it is not greater than 0.05 then the data
is not normally distributed.'''

from scipy.stats import shapiro
from scipy.stats import zscore
import matplotlib.pyplot as plt
y = np.random.normal(loc = 0, scale = 0.1, size = 100)
num_outlier = 0
for i in range(len(y_z_score)):
    if abs(y_z_score[i]) >=3:
        num_outlier+=1
shap_w, shap_p = shapiro(y)
plt.boxplot(y)
if shap_p > 0.05:
    title = 'Normally Distributed with {} outliers'.format(num_outlier)
else:
    title = 'Normally not distributed with {} outliers'.format(num_outlier)
plt.title(title)
plt.show()
```

## Exercise 18: Scatterplot

```python
'''Generate the list of numbers representing the height and save it as y'''

y = [5,5.5,5,5.5,6,6.5,6,6.5,7,5.5,5.25,6,5.25]

'''Generate a list of numbers representing weight and save it as x using the following'''

x = [100,150,110,140,140,170,168,165,180,125,115,155,135]

plt.scatter(x,y)
plt.xlabel('weight')
plt.ylabel('Height')
from scipy.stats import pearsonr
correlation_coeff, p_value = pearsonr(x,y)
print("The correlation coefficient is {}".format(correlation_coeff))
print("The p_value is {}".format(p_value))
'''The PEarson correlation coefficient is an indicator of the strength and direction of the linear relationship between
2 continuous arrays'''

if correlation_coeff == 1.0:
    title = 'There is a perfect positiove linear relationship (r = {0:0.2f})'.format(correlation_coeff)
    
elif correlation_coeff >= 0.8:
    title = 'There is a very strong positive linear relationship (r = {0:0.2f})'.format(correlation_coeff)
elif correlation_coeff >=0.6:
    title = 'There is a strong positive linear relationship (r = {0:0.2f})'.format(correlation_coeff)
elif correlation_coeff >=0.4:
    title = 'There is a moderate positive linear relationship (r = {0:0.2f})'.format(correlation_coeff)
elif correlation_coeff >=0.2:
    title = 'There is a weak positive linear relationship (r = {0:0.2f})'.format(correlation_coeff)
elif correlation_coeff >0:
    title = 'There is a very weak positive linear relationship (r = {0:0.2f})'.format(correlation_coeff)
elif correlation_coeff ==0:
    title = 'There is no linear relationship (r = {0:0.2f})'.format(correlation_coeff)
elif correlation_coeff >= -0.8:
    title = 'There is a very strong negative linear relationship (r = {0:0.2f})'.format(correlation_coeff)
elif correlation_coeff >= -0.6:
    title = 'There is a strong negative linear relationship (r = {0:0.2f})'.format(correlation_coeff)
elif correlation_coeff >= -0.4:
    title = 'There is a moderate negative linear relationship (r = {0:0.2f})'.format(correlation_coeff)
elif correlation_coeff >= -0.2:
    title = 'There is a weak negative linear relationship (r = {0:0.2f})'.format(correlation_coeff)
else:
    title = 'There is a very weak negative linear relationship(r = {0:0.2f})'.format(correlation_coeff)
plt.title(title)
plt.show()
```
